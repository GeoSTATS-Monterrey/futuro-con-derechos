## Identificamos que hay cerca de los puntos de femenicidio

```python
from shapely.ops import nearest_points
```

```python
nearest_points_list = []
for idx, row in casos_mujeres_total.iterrows():
    nearest_point = nearest_points(row['geometry'], denue_2.unary_union)
    nearest_points_list.append(nearest_point)
    
```

```python
nearest_points_list
```

```python
casos_mujeres_total['prueba_nearest'] = nearest_points_list
```
