```python
import pandas as pd
import numpy as np

df = pd.read_excel('data/variables_completas.xlsx')

df
```

```python
correlation_matrix = df.corr()

correlation_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the size for the plot
plt.figure(figsize=(15, 15))
# Square each value
correlation_matrix = np.square(correlation_matrix)

# Use seaborn to create a heatmap of the correlation matrix cmap inverted grey
sns.heatmap(correlation_matrix, xticklabels=False, yticklabels=False, cmap='gray_r')

# Show the plot
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
corr_pairs = correlation_matrix.stack()

# Exclude self-correlations (correlation of variables with themselves) which are always 1
corr_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]

# Sorting the pairs by correlation, from highest to lowest
sorted_corr_pairs = corr_pairs.sort_values(ascending=False)

# Filtering for high correlations; here we can choose a threshold, for example, 0.9
high_correlation_threshold = 0.9
high_corr_pairs = sorted_corr_pairs[sorted_corr_pairs > high_correlation_threshold]

# Filtering for low correlations; here we can choose a threshold, for example, 0.1
low_correlation_threshold = 0.1
low_corr_pairs = sorted_corr_pairs[sorted_corr_pairs < low_correlation_threshold]

# Printing out the results
print('High correlation pairs:')
print(high_corr_pairs)
```

```python
# Sum all correlations for each variable
corr_sums = correlation_matrix.sum()

# Sort the correlations from highest to lowest
sorted_corr_sums = corr_sums.sort_values(ascending=False)

sorted_corr_sums.tail(10)
```

```python
# Plot using matplotlib the sorted_corr_sums
plt.figure(figsize=(15, 10))
plt.bar(sorted_corr_sums.index, sorted_corr_sums)
plt.xticks(rotation=90)
plt.title('Sum of Correlations')
plt.show()
```
