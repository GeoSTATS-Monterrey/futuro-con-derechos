```python
import pandas as pd
import numpy as np
ruta_archivo = r'D:\Python\Python works\data\variables.csv'

df = pd.read_csv(ruta_archivo)

df.dtypes
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
variables = df.columns
variables
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the size for the plot
plt.figure(figsize=(15, 15))

# Normalize the matrix
correlation_matrix = np.square(correlation_matrix)

# Use seaborn to create a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, xticklabels=False, yticklabels=False, cmap='gray')

# Show the plot
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
corr_pairs = correlation_matrix.stack()

# Exclude self-correlations (correlation of variables with themselves) which are always 1
corr_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]

# Sorting the pairs by correlation, from highest to lowest
sorted_corr_pairs = corr_pairs.sort_values(ascending=False)

# Filtering for high correlations; here we can choose a threshold, for example, 0.9
high_correlation_threshold = 0.9
high_corr_pairs = sorted_corr_pairs[sorted_corr_pairs > high_correlation_threshold]

# Filtering for low correlations; here we can choose a threshold, for example, 0.1
low_correlation_threshold = 0.1
low_corr_pairs = sorted_corr_pairs[sorted_corr_pairs < low_correlation_threshold]

high_corr_pairs.head(20)
```

```python
# Sum all correlations for each variable
corr_sums = correlation_matrix.sum()

# Sort the correlations from highest to lowest
sorted_corr_sums = corr_sums.sort_values(ascending=False)

sorted_corr_sums.tail(10)
```

```python
# Plot using matplotlib the sorted_corr_sums
plt.figure(figsize=(15, 10))
plt.bar(sorted_corr_sums.index, sorted_corr_sums)
plt.xticks(rotation=90)
plt.title('Sum of Correlations')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import HDBSCAN
```

```python
df = df.dropna()
df
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
n_components = 50
# Punto 3: Análisis de Componentes Principales (PCA)
pca = PCA(n_components)
principal_components = pca.fit_transform(scaled_data)
```

```python
principal_components
```

```python
loading_vector = pca.components_[49]
loading_vector_abs = abs(loading_vector)
loading_vector_normalized = loading_vector_abs / loading_vector_abs.sum()
```

```python
hdbscan_model = HDBSCAN(min_cluster_size=5, min_samples=None, cluster_selection_epsilon=0.5, max_cluster_size=None, metric='euclidean', metric_params=None, alpha=1.0, algorithm='auto', leaf_size=40, n_jobs=None, cluster_selection_method='eom', allow_single_cluster=False, store_centers=None, copy=False)
```

```python
hdbscan_model.fit(principal_components)
```

```python
cluster_labels = hdbscan_model.labels_
probabilities = hdbscan_model.probabilities_
```

```python
df_with_clusters = pd.DataFrame(principal_components, columns=[f'PC{i+1}' for i in range(n_components)])
df_with_clusters['Cluster'] = cluster_labels
```

```python
print("Cluster Labels:", cluster_labels)
print("Probabilities:", probabilities)
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x=principal_components[:, 0], y=principal_components[:, 1], hue=cluster_labels, palette='viridis', s=50, edgecolors='k')
plt.title('HDBSCAN Clustering Results')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster Label', loc='upper right')
plt.show()
```
